```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
print(iris['DESCR'])
```

```python
print(iris['target'])
```

```python
iris['data']
```

```python
# x is independent feature 
X = pd.DataFrame(iris['data'],columns=['sepal length','sepal width','petal length','petal width'])
```

```python
# y is dependent feature
y = iris['target']
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
# apply decision tree classifier
from sklearn.tree import DecisionTreeClassifier
treeClassifier = DecisionTreeClassifier()
```

```python
treeClassifier.fit(x_train,y_train)
```

```python
# Visualise Decision Tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeClassifier,filled=True)
```

```python
# To avoid Overfiiting in Descion Tree :
# m1: Pre Pruning
# m2: Post Pruning
```

```python
from sklearn.tree import DecisionTreeClassifier
treeClassifier = DecisionTreeClassifier(max_depth=2)
```

```python
treeClassifier.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeClassifier)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeClassifier,filled=True) 
```

```python
# Predicting on Tree Classifier
y_pred = treeClassifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report 
cm = confusion_matrix(y_test,y_pred)
cr = classification_report(y_test,y_pred)
print(cm)
print(cr)
```

```python
# Desion Tree Pre Pruning and HyperParameter Prunning
param = {
    'criterion': ['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']
}
```

```python
param
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treemodel = DecisionTreeClassifier()
```

```python
grid = GridSearchCV(treemodel,param_grid=param,cv=5,scoring='accuracy')
```

```python
import warnings
warnings.simplefilter('ignore')
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
cm = confusion_matrix(y_test,y_pred)
cr = classification_report(y_test,y_pred)
print(cm)
print(cr)
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test,y_pred)
```
